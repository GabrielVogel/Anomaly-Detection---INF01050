```python
from datasets import load_dataset

ds = load_dataset("tanganke/eurosat")
```

```python
train = ds["train"]
test = ds["test"]
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
X_train, X_val, y_train, y_val = train_test_split(train["image"], train["label"], test_size=0.2, random_state=42,stratify=train["label"])
```

```python
X_test,y_test = test["image"],test["label"]
```

```python

from os import mkdir,listdir,remove,rmdir
```

```python
class_idx = {k:0 for k in np.unique(y_train)}
for (im,label) in zip(X_train,y_train):
    ls = listdir("train")
    if str(label) not in ls:
        mkdir(f"train/{label}")
    ls_ = listdir(f"train/{label}")

    im.save(f"train/{label}/{class_idx[label]}.png")
    class_idx[label] += 1
```

```python
class_idx = {k:0 for k in np.unique(y_val)}
for (im,label) in zip(X_val,y_val):
    ls = listdir("val")
    if str(label) not in ls:
        mkdir(f"val/{label}")
    ls_ = listdir(f"val/{label}")

    im.save(f"val/{label}/{class_idx[label]}.png")
    class_idx[label] += 1
```

```python
class_idx = {k:0 for k in np.unique(y_test)}
for (im,label) in zip(X_test,y_test):
    ls = listdir("test")
    if str(label) not in ls:
        mkdir(f"test/{label}")
    ls_ = listdir(f"test/{label}")

    im.save(f"test/{label}/{class_idx[label]}.png")
    class_idx[label] += 1
```
